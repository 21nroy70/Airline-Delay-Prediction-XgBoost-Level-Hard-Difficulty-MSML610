```python
import kagglehub, os, shutil
from pathlib import Path
raw_dir = "../data/raw"
os.makedirs(raw_dir, exist_ok=True)
path = kagglehub.dataset_download("usdot/flight-delays")
print("Kaggle path:", path)
# Copy the three CSVs if present
for fname in ["airlines.csv","airports.csv","flights.csv"]:
    src = Path(path)/fname
    if src.exists():
        shutil.copy2(src, Path(raw_dir)/fname)
print("Raw files in", raw_dir, ":", os.listdir(raw_dir))
```

```python
# Start Spark
from pyspark.sql import SparkSession
spark = (SparkSession.builder.appName("airline-delay-colab").getOrCreate())
spark.version
```
