```python
# find and cd to the repo root no matter where the notebook is opened from
from pathlib import Path
import os, sys

def find_repo_root(start: Path = Path.cwd()):
    for p in [start, *start.parents]:
        if (p / "requirements.txt").exists() and (p / "src").exists():
            return p
    return start  # fallback

ROOT = find_repo_root()
os.chdir(ROOT)
print("Project root:", ROOT)
```
